# Data Preprocessing

```python
labels = ['COVID19','NORMAL','PNEUMONIA','TURBERCULOSIS']
```

```python
X_train = []
y_train = []
image_size = 224
for i in labels:
    folderPath = os.path.join('/content/drive/MyDrive/webapp/Dataset','train',i)
    for j in tqdm(os.listdir(folderPath)):
        img = cv2.imread(os.path.join(folderPath,j))
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        img = cv2.resize(img,(image_size, image_size))
        X_train.append(img)
        y_train.append(i)
        
for i in labels:
    folderPath = os.path.join('/content/drive/MyDrive/webapp/Dataset','test',i)
    for j in tqdm(os.listdir(folderPath)):
        img = cv2.imread(os.path.join(folderPath,j))
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        img = cv2.resize(img,(image_size,image_size))
        X_train.append(img)
        y_train.append(i)
        
X_train = np.array(X_train)
y_train = np.array(y_train)
```

# SAMPLE IMAGE FROM EACH LABEL

```python
k=0
fig, ax = plt.subplots(1,4,figsize=(20,20))
fig.text(s='Sample Image From Each Label',size=18,fontweight='bold',
             fontname='monospace',y=0.62,x=0.4,alpha=0.8)
for i in labels:
    j=0
    while True :
        if y_train[j]==i:
            ax[k].imshow(X_train[j])
            ax[k].set_title(y_train[j])
            ax[k].axis('off')
            k+=1
            break
        j+=1
```

```python
X_train, y_train = shuffle(X_train,y_train, random_state=101)
```

```python
X_train.shape
```

# Dividing the dataset as a Train & Test

```python
X_train,X_test,y_train,y_test = train_test_split(X_train,y_train, test_size=0.1,random_state=101)
```

## Performing One Hot Encoding on the labels after converting it into numerical values

```python
y_train_new = []
for i in y_train:
    y_train_new.append(labels.index(i))
y_train = y_train_new
y_train = tf.keras.utils.to_categorical(y_train)


y_test_new = []
for i in y_test:
    y_test_new.append(labels.index(i))
y_test = y_test_new
y_test = tf.keras.utils.to_categorical(y_test)
```

# Building EfficientNet

```python
efficientNet = EfficientNetB0(weights='imagenet',include_top=False,input_shape=(image_size,image_size,3))
```

```python
efficientNet.summary()
```

# Visualize of the EfficientNet

```python
visualkeras.layered_view(efficientNet)
```

```python
model = efficientNet.output
model = tf.keras.layers.GlobalAveragePooling2D()(model)
model = tf.keras.layers.Dropout(rate=0.5)(model)
model = tf.keras.layers.Dense(numberOfClass,activation='softmax')(model)
model = tf.keras.models.Model(inputs=efficientNet.input, outputs = model)
```

## Layers of the Model

```python
model.summary()
```

# Visualize of the Model

```python
visualkeras.layered_view(model)
```

# Compiling the Model

```python
model.compile(loss='categorical_crossentropy',
              optimizer = 'Adam', 
              metrics= ['accuracy'])
```

```python
tensorboard = TensorBoard(log_dir = 'logs')
checkpoint = ModelCheckpoint("/content/drive/MyDrive/webapp/Model/EfficientNet0.hdf5",monitor="val_accuracy",save_best_only=True,mode="auto",verbose=1)
reduce_lr = ReduceLROnPlateau(monitor = 'val_accuracy', 
                              factor = 0.3, 
                              patience = 2, 
                              min_delta = 0.001,
                              mode='auto',verbose=1)
```

# Training the Model

```python
hist = model.fit(X_train,
                 y_train,
                 validation_split=0.1, 
                 epochs =5, 
                 verbose=1, 
                 batch_size=32,
                  callbacks=[tensorboard,checkpoint,reduce_lr]
                 )
```

# Accuracy of the Model

```python
results = model.evaluate(X_train,y_train)
```

```python
results2 = model.evaluate(X_test,y_test)
```

# CLASSIFICATION REPORT

```python
pred = model.predict(X_test)
pred = np.argmax(pred,axis=1)
y_test_new = np.argmax(y_test,axis=1)
```

```python
print(classification_report(y_test_new,pred))
```
